```python
import matplotlib.pyplot as plt

import numpy as np

from sortviz import utils as m_utils
from sortviz import sorters as m_sorters
```

```python
num_rows = 12
num_cols = 10
np.random.seed(42)
data = m_utils.generate_data(num_rows, num_cols)
```

# Generate the Data to Sort

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.imshow(data)
ax.set_title("Random Data to Sort")
ax.set_ylabel("Row (#)")
ax.set_xlabel("Col (#)")
```

# Bubble Sort

## Sort Data

```python
bubble_sorter = m_sorters.BubbleSorter()
bubble_sorted_data = bubble_sorter.sort(data.copy())
bubble_sorter.save_sort("bubble_sort.json")
```

## Plot Sort Passes

```python
bubble_figs = m_utils.plot_sort("bubble_sort.json")
for idx, fig in enumerate(bubble_figs):
    fig.savefig(f"../../images/bubble_sorter_{idx}.png")
```

# Merge Sort

## Sort Data

```python
merge_sorter = m_sorters.MergeSorter()
merge_sorter_data = merge_sorter.sort(data.copy())
merge_sorter.save_sort("merge_sort.json")
```

## Plot Iterations

```python
merge_figs = m_utils.plot_sort("merge_sort.json")
for idx, fig in enumerate(merge_figs):
    fig.savefig(f"../../images/merge_sorter_{idx}.png")
```

# Quick Sort

## Sort Data

```python
quick_sorter = m_sorters.QuickSorter()
quick_sorter_data = quick_sorter.sort(data.copy())
quick_sorter.save_sort("quick_sort.json")
```

## Plot Iterations

```python
quick_figs = m_utils.plot_sort("quick_sort.json")
for idx, fig in enumerate(quick_figs):
    fig.savefig(f"../../images/quick_sorter_{idx}.png")
```

# Compare Sorters

```python
comp_fig = m_utils.plot_sort_comparison(
    ["bubble_sort.json", "merge_sort.json", "quick_sort.json"], figsize=(10, 10)
)
comp_fig.savefig("../../images/sort_comparison.png")
```
